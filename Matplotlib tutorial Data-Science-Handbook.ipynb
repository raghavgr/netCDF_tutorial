## Visualization with Matplotlib

### 1. Intro

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
x
```

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x)) # figures that are later displayed with plt.show()

plt.show()
# Use plt.draw() to force plot display if plt.show() doesn't work.
```

```python
fig = plt.figure()
plt.plot(x, np.sin(x), dashes=[3,1,10,2])
plt.plot(x, np.cos(x), '+')
```

```python
fig.savefig('my_figure.png')
```

```python
plt.figure() # create a plot figure

# create the first of four panels and set current axis
plt.subplot(2, 2, 1) 
plt.plot(x, np.sin(x), dashes=[4, 2, 10, 2])
plt.subplot(2, 2, 2)
plt.plot(x, np.tan(x), dashes=[4, 2, 10, 2], color='red')
plt.subplot(2, 2, 3)
plt.plot(x, np.cos(x), dashes=[5, 1, 11, 1], color='green')
plt.subplot(2, 2, 4)
plt.plot(x, np.sin(-x), dashes=[4, 1, 10, 2], color='magenta')
```

```python
fig, ax = plt.subplots(2, 2, sharex=True)
functions = [np.sin(x), np.sin(-x), np.tan(x), np.cos(x)]
titles = ['sin(x)', 'sin(-x)', 'tan(x)', 'cos(x)']

# use zip instead of enumerate to iterate over multiple arguments
# and use axes.flat
for ax, title, function in zip(ax.flat, titles, functions):
    ax.plot(x, function)
    ax.set_title(title)
    
```

```python
# First create a grid of plots
# ax will be an array of two Axes objects
fig, ax = plt.subplots(2)

# Call plot() method on the appropriate object
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x));
```
